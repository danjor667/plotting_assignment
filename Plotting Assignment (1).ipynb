# Histograms 

```python
carbon_dioxide = data['co2']
plt.figure(figsize=(10,8))
plt.hist(carbon_dioxide, bins=30, edgecolor='black')
plt.title('Carbon Dioxide Distribution')
plt.yticks(np.arange(0, 450, 50))
plt.xlabel('Carbon Dioxide')
plt.ylabel('Frequency')
# plt.yscale('log')
plt.show()
```

# plotting the Distribution of pv2 and pv4

```python
# creating grid for the 2 diagrams
gs = GridSpec(2, 2, height_ratios=[1, 1], width_ratios=[1, 1])
plt.figure(figsize=(10,8))
ax1 = plt.subplot(gs[0, :])
ax2 = plt.subplot(gs[1, :])

# fetching corresponding column data
pv2 = data['pv2']
pv4 = data['pv4']

# plotting fig
ax1.hist(pv2, bins=30, edgecolor='black')
ax2.hist(pv4, bins=30, edgecolor='black')
ax1.set_title('pv2 Distribution')
plt.ylabel("Frequency")
plt.xlabel('pv4 Distribution')
plt.xticks(np.arange(0, 130, 10))
plt.show()
```

```python
# function to remove all the non numerical columns 
def remove_colums(df):
    """
    return dataset with no non numerical value
    :param df: input dataFrame
    :return: clean dataFrame
    """
    return df.select_dtypes(include=["number"])

data = remove_colums(data)
data.head()
```

```python
correlation_matrix = data.corr()
plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, vmin=-1, vmax=1, cmap='vlag')
plt.title('Correlation Heat Map')

plt.show()
```
